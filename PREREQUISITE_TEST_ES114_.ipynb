**2. Mutable vs. Immutable Data Types:**

Explain the difference between mutable and immutable data types in Python. Provide examples of each.

## Mutable vs. Immutable Data Types in Python



**Immutable Data Types:**

* **Definition:** Immutable objects cannot be modified after creation. Any operation that appears to modify an immutable object actually creates a new object with the modified value.
* **Examples:** `int`, `float`, `str`, `tuple`, `bool`, `frozenset`
*  When you perform an operation that seems to modify an immutable object, a new object is created with the new value, and the original object remains unchanged.

**Mutable Data Types:**

* **Definition:** Mutable objects can be modified after creation. Operations performed on these objects directly change their values without creating new objects.
* **Examples:** `list`, `dict`, `set`
 When you modify a mutable object, the changes are reflected in the original object itself.

**Key Differences:**

| Feature | Mutable | Immutable |
|---|---|---|
| Modifiability | Can be changed after creation | Cannot be changed after creation |
| Object Identity | Changes affect the original object | Changes create a new object |


**Example:**

```python
#Immutable Data Types:
my_tuple = (1, 2, 3)
my_tuple[0] = 4
```

```python
#Immutable Data Types:
text = "hello"
new_text = text.upper()
print(text)
print(new_text)
```

```python
# Mutable Data Types:
# Lists

my_list = [1, 2, 3]
my_list.append(4)
print(my_list)

#Dictionaries

my_dict = {'a': 1, 'b': 2}
my_dict['c'] = 3
print(my_dict)

#Sets

my_set = {1, 2, 3}
my_set.add(4)
print(my_set)
```

**3.Fibonacci Sequence:**

* Write a Python function that takes a number n and returns a list of the first n Fibonacci numbers.





* Plot the Fibonacci sequence using a line graph.

```python


def fib(num):
  ls=[0,1]
  if num==0 or num==1:

    print(ls[0])
  elif num==2:

    print(ls)

  else:
    for i in range(2,num):
      ls.append(ls[i-1]+ls[i-2])
  return(ls)

print("enter an integer ")
num=int(input())
ans=fib(num)
print(f"Fibonacci series is : {ans} ")


import matplotlib.pyplot as plt
plt.plot(ans, marker='o')
plt.title("Fibonacci Sequence")
plt.xlabel("Term")
plt.ylabel("Value")
plt.xticks(range(num))
plt.yticks(range(min(ans),max(ans)+1))
plt.xlim(0, num-1)
plt.ylim(0, max(ans) )

plt.show()




```

 * Count how many strings start with a vowel.

```python
def longest_string(strings):
  longest=""
  for i in strings:
    if len(i)>len(longest):
      longest=i
  print(f"Longest string is: {longest}" )

def start_with_vowel(strings):
  vowels=['a','e','i','o','u','A','E','I','O','U']
  count=0
  for i in strings:
    if i[0] in vowels:
      count+=1
  print(f"Number of strings starting with a vowel is : {count}")



ls=['string','is','nothing','but','list','of','characters','in','python']
longest_string(ls)
start_with_vowel(ls)

import matplotlib.pyplot as plt
length_list=[len(i) for i in ls]
plt.bar(ls, length_list)
plt.xlabel('String')
plt.ylabel(' Length of String ')
plt.title('String Length Distribution')
plt.show()


```

**17. File Operations:**

Write Python code to:
* Create a text file.
* Write a list of numbers to the file (one per line).
* Read the file and compute their sum.

```python
def write_numbers_to_file(file_name, numbers):

    try:
        with open(file_name, 'w') as file:
            for number in numbers:
                file.write(str(number) + '\n')
        print(f"Numbers written to '{file_name}' successfully.")

    except Exception as e:
        print(f"Error writing to file: {e}")

def read_numbers_and_sum(file_name):

    total_sum = 0
    try:
        with open(file_name, 'r') as file:
            print(f"\nContents of '{file_name}':")
            for line in file:
                number = int(line.strip())
                total_sum += number
                print(number)
            print(f"Sum of numbers in '{file_name}': {total_sum}")
    except FileNotFoundError:
        print(f"File '{file_name}' not found.")
    except ValueError:
        print(f"Invalid data in file. Make sure all lines contain numbers.")

numbers_list = [10, 25, 5, 18, 32]
file_name = "numbers.txt"

write_numbers_to_file(file_name, numbers_list)
read_numbers_and_sum(file_name)
```

## 21. Basic Statistics:
- Write Python code to calculate:
 - Mean, median, mode, and standard deviation of a list of numbers.
 - Visualize the data using a box plot.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def calculate_stats(data):

    mean = np.mean(data)
    median = np.median(data)
    mode = stats.mode(data)
    std_dev = np.std(data)

    return mean, median, mode, std_dev

def visualize_data(data):

    plt.figure(figsize=(8, 6))
    plt.boxplot(data)


    plt.title("Box Plot of Data")
    plt.xlabel("Value")
    plt.show()


data = [12, 15, 18, 20, 22, 25, 25, 28, 30, 35, 38, 42]

mean, median, mode, std_dev = calculate_stats(data)
print(f"Mean: {mean:.2f}")
print(f"Median: {median:.2f}")
print(f"Mode: {mode:}")
print(f"Standard Deviation: {std_dev:.2f}")

visualize_data(data)
```

## 22. Normalization:
- Write a Python function to normalize a list of numbers to a range of 0 to 1.
- Visualize the normalized data using a line plot.

```python
import matplotlib.pyplot as plt
import random

num = [random.randint(1,100) for _ in range(10)]
print(" Original list of Numbers: ",num)

def normalize_num(numbers):
  min_num = min(numbers)
  max_num = max(numbers)
  normalized_numbers = [(x - min_num) / (max_num - min_num) for x in numbers]
  return normalized_numbers



ans = normalize_num(num)
print("Normalized numbers in range 0 to 1 are : ",ans )

plt.plot(ans,marker='o',linestyle='-',color='red')
plt.xlabel('Index')
plt.ylabel('Values')
plt.title('Normalized Data')
plt.grid(True)
```
